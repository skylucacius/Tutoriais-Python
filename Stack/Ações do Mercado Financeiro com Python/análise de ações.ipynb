```python
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
from statsmodels.tsa.seasonal import seasonal_decompose
import pandas as pd
```

# Decomposição

```python
ticket = yf.Ticker('^bvsp')
```

```python
df = ticket.history('1day', start='2018-01-01', end='2021-01-01')
```

```python
decomposicao = seasonal_decompose(df['Close'], model='additive',period=30, extrapolate_trend=30)
```

```python
decomposicao_multiplicativa = seasonal_decompose(df['Close'], model='multiplicative', period=30, extrapolate_trend=30)
```

```python
decomposicao.trend + decomposicao.seasonal + decomposicao.resid == decomposicao.observed
```

```python
decomposicao_multiplicativa.trend * decomposicao_multiplicativa.seasonal * decomposicao_multiplicativa.resid == decomposicao_multiplicativa.observed
```

```python
decomposicao.plot()
```

```python
decomposicao_multiplicativa.plot()
```

# Média Móvel

```python
df['Close'].rolling(7).mean()
```

```python
mediamovel_7dias = df['Close'].rolling(7).mean()
mediamovel_14dias = df['Close'].rolling(14).mean()
mediamovel_21dias = df['Close'].rolling(21).mean()
```

```python
fig, ax = plt.subplots(figsize=(12,5))
plt.plot(mediamovel_7dias,'orange')
plt.plot(mediamovel_14dias,'red')
plt.plot(mediamovel_21dias,'blue')
plt.plot(df['Close'])

```
