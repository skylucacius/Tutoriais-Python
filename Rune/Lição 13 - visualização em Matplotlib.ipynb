```python
import csv
filename = "files/cars.csv"
with open(filename) as file:
    spamreader = csv.DictReader(file)
    lista = list(spamreader)

    hp = [int(linha["Horsepower"]) for linha in lista]
    torque = [int(linha["Torque"]) for linha in lista]
    fuel_city_mpg = [int(linha["Fuel City mpg"]) for linha in lista]

    import matplotlib.pyplot as plt
    fig, ax = plt.subplots()
    ax.scatter(hp, torque, c=fuel_city_mpg, alpha=.05)
```

```python
import random

def roll_dice():
    return random.randint(1,6) + random.randint(1,6)

# quanto maior a quantidade de interaÃ§Ãµes, mais o histograma se aproxima de uma forma "triangular" ...
# ... para isso, basta adicionar alguns zeros e visualizar o resultado.
lista = [roll_dice() for i in range(500)]

import matplotlib.pyplot as plt
ax, fig = plt.subplots()
fig.hist(lista, bins=11)
```
